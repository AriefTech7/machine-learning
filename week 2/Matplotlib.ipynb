# Line Plot
digunakan untuk:
1. Melihat perubahan nilai dari waktu to waktu, example app trading
2. Time series data

```python
x = [1, 2, 3, 4, 5]
y = [10, 35, 20, 75, 80]

plt.figure(figsize=(8, 4))
# plt.plot()
plt.plot(x, y)
plt.title("example line plot")
plt.ylabel('sumbu Y')
plt.xlabel('sumbu X')
plt.grid(True)
plt.show()# bisa digunakan ketika menggunakan notebooks
```

```python
x = [1, 2, 3, 4, 5]
y2 = [15, 20, 30, 25, 15]
y = [10, 35, 20, 75, 80]

plt.figure(figsize=(8, 4))
# plt.plot()
plt.plot(x, y, marker='o', color='red', linestyle='--')
# plt.fill(x, y, color='purple', label='Area Terarsir', alpha=0.5)
plt.plot(x, y2, marker='o', color='blue')


plt.title("Mata Kuliah")
plt.ylabel('sumbu Y')
plt.xlabel('sumbu X')
plt.grid(True)
plt.legend(['Jaringan', 'B.inggris'])
plt.savefig('Matkul') 

plt.show()# bisa digunakan ketika menggunakan notebooks
```

## Bar Plot
digunakan untuk:
1. Membandingkan nilai

```python
kategori = ['python', 'Assembly', 'Java', 'C++']
jumlah_user= [100, 10, 50, 70]

plt.figure(figsize=(9, 5))
# plt.bar
plt.bar(kategori, jumlah_user, color='purple')

# element
plt.title('Pengguna Bahasa pemrograman', fontweight='bold')
plt.xlabel('Bahasa Pemgrograman')
plt.ylabel('Jumlah User')

plt.grid(axis='y', alpha=0.5, linestyle= '--')
plt.show()
```

```python
kategori = ['Batch 1', 'Batch 2', 'Batch 3']
laki_laki = [40, 60, 55]
perempuan = [15, 30, 50]

# menentukan ukuran each bar
x = np.arange(len(kategori))
bar_with= 0.35

plt.figure(figsize=(8, 5))
# plt.bar
plt.bar(x - bar_with/2, laki_laki, width=bar_with)
plt.bar(x + bar_with/2, perempuan, width=bar_with)

# element
plt.title('Jumlah Student Class ML', fontweight='bold')
plt.xlabel('Jumlah Batch')
plt.ylabel('Jumlah Student')

plt.xticks(x, kategori)# memberikan nama untuk setiap bar
plt.grid(axis='y', alpha=0.5, linestyle= '--')
plt.legend(['Laki laki', 'Perempuan'])
# menata layout agar tidak terpotong
# plt.tight_layout()
# ga boleh dideklarasi sesudah plt.show()
plt.savefig('Class ML')
plt.show()

```

## Scatter Plot
Digunakan untuk:
1. Menampilkan hubungan beetwen 2 variable numerik atau lebih

```python
umur = [18, 21, 25, 30, 35, 40, 45, 50]
penghasilan = [1.5, 2.2, 2.8, 4.0, 4.2, 4.5, 4.7, 4.9]

plt.figure(figsize=(8, 5))
# plt.scatter()
plt.scatter(umur, penghasilan)

plt.title('Hubungan antara umur dan gaji')
plt.xlabel('Umur')
plt.ylabel('penghasilan(Juta rupiah)')

plt.grid(True)
plt.tight_layout()
plt.legend(['penghasilan'])
plt.savefig('Hubungan antara umur dan gaji')
plt.show()
```

```python
fisika_A = [75, 80, 85, 90, 78]
matematika_A = [70 ,85, 82, 90, 99]

fisika_B = [60, 65, 61, 20, 90]
matematika_B = [65, 62, 70, 80, 11]

fisika_C = [70, 60, 61, 20, 90]
matematika_C = [65, 50, 50, 80, 11]

plt.figure(figsize=(8, 5))
# plt.scatter()
plt.scatter(fisika_A, matematika_A, marker='*', label='nilai kelas A', color='blue')
plt.scatter(fisika_B, matematika_B, marker='o', label ='nilai kelas B', color='red')
# plt.scatter(fisika_C, matematika_C, label='nilai kelas C', marker='X')

plt.title('hubungan antara MTK & Fisika')
plt.xlabel('Fisika')
plt.ylabel('Matematika')

plt.grid(True, alpha= 0.5)
plt.tight_layout()
plt.legend(loc='lower left')
plt.savefig('hubungan antara MTK & Fisika')
plt.show()


```

```python
# plt.boxplot


plt.figure(figsize=(10, 6))
# param vert berfungsi untuk membuat box plot menjadi vertical
plt.boxplot(df['Total Amount'], vert=False)


plt.title("Boxplot Total Amount Retail Sales Data", fontsize=20)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Customer", fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```

ini bisa menjadi salah satu method untuk melihat outliars

```python
views = [320, 450, 400, 390, 410, 470, 500, 490, 480, 510,
         530, 600, 620, 590, 550, 570, 610, 640, 700, 2000, 3000, 4000, 5000]

plt.figure(figsize=(10, 6))
plt.boxplot(views, vert=False)


plt.title("Outliars pada data views", fontsize=20)
plt.xlabel("Number of views", fontsize=12)
plt.ylabel("tiktok", fontsize=12)
plt.grid(axis="y", alpha=0.3)
plt.show()
# titik-titik pada gambar dibawah adalah outliar
```

# Subplots

```python
hari = ['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min']
followers = [120, 130, 145, 160, 175, 190, 210]
views = [300, 400, 350, 500, 600, 550, 700]
likes = [100, 120, 110, 160, 200, 180, 250]
comments = [20, 25, 22, 35, 40, 38, 50]
engagement_rate = [0.4, 0.36, 0.38, 0.39, 0.4, 0.39, 0.43]

# plt.subplots

fig, axs = plt.subplots(2, 2, figsize=(12, 8)) # untuk membuat subplots
axs[0, 1].plot(hari, followers, marker='o')
axs[0, 1].set_title("Pertumbuhan Follower")
axs[0,1].set_ylabel("Jumlah Follower")
axs[0,1].grid(alpha=1)

axs[0,0].plot(hari, views, marker='o')
axs[0,0].set_title("Pertumbuhan Views")
axs[0,0].set_ylabel("Jumlah Views")
axs[0,0].grid(alpha=1)

axs[1,0].bar(hari, likes, color='Skyblue')
axs[1,0].set_title("Pertumbuhan Like")
axs[1,0].set_ylabel("Jumlah Likes")
axs[1,0].grid(alpha=1)

axs[1,1].scatter(hari, engagement_rate, color='purple')
axs[1,1].set_title("Pertumbuhan engagement rate")
axs[1,1].set_ylabel("Jumlah engagement rate")
axs[1,1].grid(alpha=1)
axs[1,1].set_ylim(0 , 0.5)


fig.suptitle("Grafik Medsos", fontsize=20)
fig.tight_layout()
```

## Correlation Matrix
note:
harus data numerical

```python
df.head(10)
```

```python
numerical_df = df[['Quantity', 'Age', 'Price per Unit', 'Total Amount']]
# .corr untuk mencari correlation matrix
# round disini hanya membatas angka sesudah koma
corr_matrix = round(numerical_df.corr(), 2)
print(corr_matrix)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
# sns.heatmap untuk correlation matrix
sns.heatmap(
    corr_matrix,
    annot=True,      # Show correlation values on the heatmap/menampilkan nilai korelasi pada heatmap
    cmap='coolwarm', # Choose a colormap (e.g., 'coolwarm', 'viridis', 'plasma')
    linewidths=0.5,
    fmt='.2f',
    cbar_kws={'label': r'koefisies korelasi'}
)
plt.title('Correlation Matrix', fontsize=20)
plt.tight_layout()
plt.show()
# semakin merah warna semakin berkorelasi 
```
