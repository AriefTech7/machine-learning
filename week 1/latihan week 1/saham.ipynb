```python
datetime_colom = [
    'ListingDate', 'MinutesFirstAdded',
    'MinutesLastUpdated','HourlyFirstAdded', 
    'HourlyLastUpdated', 'DailyFirstAdded',
    'DailyLastUpdated'
]
```

```python
for col in datetime_colom:
    # mengkonversi ke tipe data datetime
    dataframe[col] = pd.to_datetime(dataframe[col], errors='coerce')
    # mengisi data yang kosong menggunakan interpolate
    dataframe[col] = dataframe[col].interpolate()# interpolate -> mengisi data yg hilang dengan menebak nilai berdasarkan data sebelum dan sesudahnya
```

```python
dataframe['LastPrice'] = dataframe['LastPrice'].fillna(0)
```

```python
dataframe['MarketCap'] = dataframe['MarketCap'].fillna(0)
```

### feature enggineering

```python
# function untuk mencategorikan market cap
def categori_market(cap):
    if cap >= 10000000000000:
        return 'Big Cap'
    elif 1000000000000 <= cap < 10000000000000:
        return 'Mid Cap'
    else:
        return 'Small Cap'
```

```python
dataframe['CategoriMarketCap'] = dataframe['MarketCap'].apply(categori_market)
```

```python
categori = dataframe['CategoriMarketCap']
dataframe = dataframe.drop('CategoriMarketCap', axis=1)
dataframe.insert(4, 'CategoriMarketCap', categori)
```
