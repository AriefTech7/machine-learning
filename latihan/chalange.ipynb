```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# challange 1 line plot 
kelas_A = [55, 80, 65, 100, 130, 150]
kelas_B = [40, 70, 60, 90, 120, 140]
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']

panjang = np.arange(len(bulan))

plt.figure(figsize=(9, 5))

plt.plot(kelas_A, color='blue', marker='o', linestyle='--', label='kelas_A')
plt.plot(kelas_B, color='red', marker='s', label='kelas_B')

plt.title('Perkembangan Jumlah Peserta per Kelas')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peserta')

plt.xticks(panjang, bulan)
plt.grid(True)
plt.legend()
plt.savefig('Perkembangan Jumlah Peserta per Kelas')
plt.tight_layout()
plt.show()
```

```python
# challange 2 bar plot
bahasa_pemrograman = ['Python', 'Java', 'C++', 'Javascript']
jumlah_2024= [83, 70, 60, 76]
jumlah_2025 = [87, 72, 61, 80]

# menentukkan darimana bar dibuat
x = np.arange(len(bahasa_pemrograman))
size =0.35

plt.figure(figsize=(8, 5))
plt.bar(x- size/2, jumlah_2024, size, color='skyblue')
plt.bar(x + size/2, jumlah_2025, size, color='orange')


plt.title('Jumlah Pengguna Bahasa Pemrograman')
plt.xlabel('bahasa_pemrograman')
plt.ylabel('Jumlah(juta)')

plt.xticks(x, bahasa_pemrograman)
plt.grid(axis='y', alpha= 0.5,linestyle='--')
plt.legend(['2024', '2025'], loc='upper center')
plt.tight_layout()
plt.savefig('Jumlah Pengguna Bahasa Pemrograman')
plt.show()
```

```python
# challenge 3 scatter plot

laki_laki_tg = [165, 170, 175, 180, 185]
laki_laki_bb = [60, 65, 73, 78, 85]
perempuan_tg = [155, 160, 163, 168, 170]
perempuan_bb = [50, 53, 55, 60, 62]

plt.figure(figsize=(6, 5))

plt.scatter(laki_laki_tg, laki_laki_bb, marker='o', color='blue')
plt.scatter(perempuan_tg, perempuan_bb, marker='^', color='magenta')
plt.title('Perbandingan Tinggi dan Berat Badan')
plt.xlabel('Tinggi Badan(cm)')
plt.ylabel('Berat Badan(kg)')

plt.grid(True)
plt.tight_layout()
plt.legend(['laki laki', 'perempuan'])
plt.savefig('Perbandingan Tinggi dan Berat Badan')
plt.show()
```
